# Global Earthquake & Tsunami Tracking
## Dataset sourced from Kagle:
https://www.kaggle.com/datasets/ahmeduzaki/global-earthquake-tsunami-risk-assessment-dataset <br>
This dataset contains the seismic characteristics of earthquakes and potential indicators for tsunamis globally from 2001 to 2022. This data can have many important use-cases in our real world, whether it be for tsunami risk prediction, earthquake analysis, or seismic hazard assessment applications.

### Column Guide
magnitude = Earthquake magnitude from 6.5 - 9.1 <br>
cdi	= Community Decimal Intensity (the felt intensity) from 0 - 9 <br>
mmi	= Modified Mercalli Intensity, ranging from 1 - 9 <br>
sig	= Event significance score from 650 - 2910 <br>
nst	= Number of seismic monitoring stations	from 0 - 934 <br>
dmin =	Distance to nearest seismic station 0.0 - 17.7 <br>
gap	= Azimuthal gap between stations 0.0 - 239.0 <br>
depth =	Earthquake focal depth in km, ranging from 2.7 - 670.8 <br>
latitude = Epicenter latitude -61.85째 to 71.63째	<br>
longitude = Epicenter longitude -179.97째 to 179.66째	<br>
Year = Year of occurrence from 2001 - 2022 <br>
Month =	Month of occurrence	from 1 (January) to 12 (December) <br>
tsunami	= Tsunami potential, either 0 or 1

```python
income_categories = ["low-income", "lower-middle-income", "upper-middle-income", "high-income"]
num_complete_sets = len(df) // 4
remainder = len(df) % 4
df["inc"] = income_categories * num_complete_sets + income_categories[:remainder]
del df["new column"]
print(df.head())
```

#### This new column represents the income classification of the impacted country according to world bank data

```python
filtered_magnitude = df[df['magnitude'] > 8.0]
filtered_cdi = df[df['cdi'] > 7.0]
print(filtered_magnitude)
print(filtered_cdi)
```

#### I created these two subsets because I wanted to examine the specific instances where earthquakes had the highest magnitudes and the greatest intensities were felt by humans
