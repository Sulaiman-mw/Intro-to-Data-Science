```python
df_clean = df.copy()
print(f"Original df shape: {df.shape}")
print(f"New df_clean shape: {df_clean.shape}")
```

```python
df_clean['DATE'] = pd.to_datetime(df_clean['DATE'])
df_clean['MONTH'] = df_clean['DATE'].dt.month
print("Added MONTH column based on DATE")
print(df_clean[['DATE', 'MONTH']].head())
print(df_clean[['DATE', 'MONTH']].tail())
```

```python
df_filtered = df_clean.dropna(subset=['TAVG'])
print(f"Filtered data (non-null TAVG): {df_filtered.shape}")
```
